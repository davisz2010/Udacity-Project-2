### Which states have had the highest growth in gun registrations?

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.

# Find the total guns owned for each state

state_total = df1.groupby(['month', 'state'])['totals'].sum()
```

```python
# earliest date
start = df1['month'].min()
start
```

```python
# latest date
end=df1['month'].max()
end
```

```python
# since some states has a 0 guns at the start date, which means gun owner ship's growth percentage... 
# ...would not be an accurate measure, then we'll only measure growth by the increase in gun ownership
state_total.loc[start]
```

```python
#the highest growth of states between start and end time 
growth=(state_total.loc[end]-state_total.loc[start])
max(growth)
```

```python
growth.idxmax()

```

```python
# find the initial total of gun registrtaion for kentucky 
start_total=state_total.loc[start]
start_total.loc['Kentucky']

```

### Conclusion: Kentucky has the highest growth in gun registrations. It went from 840 to 397866.

### Research Question 2  What is the overall trend in gun purchases

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
# graph functio to save time
def graph (df,chart_title,x_axis,y_axis):
    df.plot(kind='line')
    plt.title(chart_title)
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)

```

```python
#Groupby month and total gun registration
#
x=df1.groupby('month')['totals'].sum()
# a line plot is needed so I can visually see the growth of gun purchases as time increases
graph(x,"Gun Purchase Trend by years Line Chart",'Year','Total gun purchases')


```

###  Conclusion: As can be seen, the total gun registration has been increasing annually, and it has reached its peak sometime in 2016.
